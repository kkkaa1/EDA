```python
df_subway=pd.read_csv("2020_2022지하철승하차.csv",encoding='utf8')
df_subway.head()
```

```python
df_subway.loc[:, 'sum' ] = df_subway.loc[:, '승차총승객수':'하차총승객수'].sum(axis=1)
df_subway.head()
```

```python
#9호선 1단계랑 2~3단계 통일
df_subway['노선명']=df_subway["노선명"].str.replace("2~3단계","")
df_subway.tail(30)
```

```python
print(df_subway[df_subway['노선명']=='우이신설선'])
```

```python
# 0,      1,    299,    582,    583,    584,    585,    586               587,    588,
droplst=[ '우이신설선',  '공항철도 1호선',  '경강선', '수인선', '경춘선', '경의선', '장항선', '중앙선', '일산선', '분당선',
       '과천선', '안산선', '경원선', '경인선', '경부선',  '신림선']
indexlst=[]
for i in droplst:
    indexlst.append(df_subway[df_subway['노선명']==i].index)
print(indexlst)
```

```python
def flatten(lst):
    result=[]
    for i in lst:
        result.extend(i)
    return result
indexlist=flatten(indexlst)
```

```python
df_subway=df_subway.drop(indexlist)
df_subway.head()
```

```python
df_subway.head()
```

```python
df_subway["사용일자"]=df_subway["사용일자"].str.replace("2020","2020-")
df_subway["사용일자"]=df_subway["사용일자"].str.replace("2021","2021-")
df_subway["사용일자"]=df_subway["사용일자"].str.replace("2020","2020-")
```
